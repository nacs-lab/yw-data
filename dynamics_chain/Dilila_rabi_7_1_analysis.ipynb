```python
import numpy as np
import matplotlib.pyplot as plt
t = np.array([1,1.5,3,5])
standard_dev = np.array([0.06821449118981772,0.07056041556025712,0.08434224572056602,0.0968587396406361])
corrected_sd= np.array([0.046490682785756514,0.0454644606871541,0.05849726129811209,0.07218124494822664])
```

```python
plt.plot(t,standard_dev,marker = 'o',label = 'Standard Deviation')
plt.plot(t,corrected_sd,marker = 'o',label = 'Standard Deviation with correction')
plt.xlabel('Dilila duration(us)')
plt.ylabel('SD of Ground state population')

plt.legend()
```
