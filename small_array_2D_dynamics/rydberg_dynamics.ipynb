```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from itertools import product
from scipy.optimize import curve_fit
import matplotlib.colors as colors
```

# benchmark with 3*5 system

```python
#ramp time =2us
data_exp = np.loadtxt('./rydberg_data/3x5Data_8_25_slow.csv',delimiter=',')
data = {}
for params in [(5,3,0.0,-4,4,0.649,2.,0.05,0)]:
    data[params] = {}
    data[params]['z'] = np.loadtxt("./rydberg_data/9_1_3x5/z_L%d_%d_den%.3f_delta%.1f_%.1f_omega%.3f_ramptime%.3f_dt%.3f_No%d.csv"%params,delimiter=',')
params = (5,3,0.0,-4,4,0.649,2.,0.05,0)
dt = params[7]
temp = data[params]['z']
fig,ax = plt.subplots(ncols=3,nrows=5,figsize=(7,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
for i in range(15):
    #print(temp[:,i])
    ax[int(i/3),i%3].errorbar(data_exp[0,:],data_exp[i+1,:],yerr=data_exp[i+1+15,:],ls='--',marker='.',label='experiment')
    ax[int(i/3),i%3].plot(np.arange(len(temp[:,i]))*dt,1-(temp[:,i]+1)/2,ls='-',label='numerics')
    #plt.legend(handlelength=0.5,ncol=3)
    #plt.text(20,0.75,'size=3*5',fontsize=25)
    ax[int(i/3),i%3].set_ylim([0,1])
    ax[int(i/3),i%3].set_xticks([0,1,2,3])
    ax[int(i/3),i%3].set_yticks([0,0.2,0.4,0.6,0.8,1])

    #ax[i/3,i%3].set_xlim([0,1])
    ax[int(i/3),i%3].grid()

    ax[int(i/3),i%3].set_xlabel('Time(us)')
    ax[int(i/3),i%3].set_ylabel('G survival')
ax[4,2].legend(handlelength=0.5)

plt.show()
```

```python
z1 = 1-(temp[-1,:]+1)/2
z1 = z1.reshape(5,3)
z2 = data_exp[1:16,-1]
z2 = z2.reshape(5,3)

x = np.arange(0, 4, 1) 
y = np.arange(0, 6, 1) 
xx,yy = np.meshgrid(x,y)

print(x)
print(y)
print(np.shape(z1))
fig, ax = plt.subplots(ncols = 2,nrows = 1,figsize=(7,3))
c0 = ax[0].pcolormesh(xx, yy, z1,norm=colors.Normalize(vmin=0, vmax=1))
ax[0].set_yticks([0,1])
c1 = ax[1].pcolormesh(xx, yy, z2,norm=colors.Normalize(vmin=0, vmax=1))
fig.colorbar(c0, ax = ax)
ax[0].set_yticks([1,3,5])
ax[0].set_xticks([0,1,2,3])
ax[0].set_title('numerics')

ax[1].set_yticks([1,3,5])
ax[1].set_xticks([0,1,2,3])
ax[1].set_title('experiments')
# #ax.set_ylim([0,1])
# ax.set_xlim([1,3])
```
