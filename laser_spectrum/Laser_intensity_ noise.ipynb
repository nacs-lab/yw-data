```python
import numpy as np
import os
file1 = '0.txt'
file2 = '1.csv'
data = {}
data['frequency'] = []
data['power1'] = []
data['power2'] = []

with open(file1,'r') as f:
    lines = f.readlines()[13:]
    for line in lines:
        line_1=line.split(';')
        #print(line_1)
        data['frequency'].append(float(line_1[0]))
        data['power1'].append(float(line_1[1]))
data['frequency'] = np.array(data['frequency'])
data['power1'] = np.array(data['power1'])
with open(file2,'r') as f:
    lines = f.readlines()[13:]
    for line in lines:
        line_1=line.split(';')
        data['power2'].append(float(line_1[1]))
data['power2'] = np.array(data['power2'])
```

```python
plt.plot(data['frequency']/1e6,data['power1'],label = 'BG')
plt.plot(data['frequency']/1e6,data['power2'],label = 'noise')
plt.legend()
FS = 18
plt.ylabel('power(dBm)',fontsize=FS)
plt.xlabel('frequency(MHz)',fontsize=FS)
```
