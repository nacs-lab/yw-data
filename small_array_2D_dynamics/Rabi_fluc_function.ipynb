```python
# random
def func_Omega(t):
    temp = []
    for i in range(len(t)):
        temp.append(Omega*((random()-0.5)*A_fluc+1))
    return np.array(temp)
```

```python
# noise at a certain frequency
def func_Omega(t):
    temp = []
    phase = random()*np.pi*2
    for i in range(len(t)):
        temp.append(Omega*(1+A_fluc*np.sin(2*np.pi*Omega*t[i]+phase)))
    return np.array(temp)
```
