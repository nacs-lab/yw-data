```python
import numpy as np
import matplotlib.pyplot as plt
N = 20
data = np.load('C:/Users/wangy/Rydberg simulation/rydberg_data/1_4_23/2023_01_04_15_41_58L20.npy',allow_pickle=True).tolist()
t = np.array(data['nT'])[:,1] # scan time
detuning = np.array(data['nT'])[:,3] # detuning
rabi = np.array(data['nT'])[:,2] # Rabi
E = np.array(data['E']) # energy for the lowest three states
overlap = np.array(data['overlap'])# overlap with fm, afm, ground state, 1st excited state, 2nd excited state


data_d = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/1_4_23/Overlap2023_01_04_01_39_11_L20_1',delimiter=',')

```

```python
ms = 2
ls = '-'
fig, ax = plt.subplots(2, 3, figsize=(15,10))
fig.suptitle('10 Atoms')
plt.subplots_adjust(wspace=0.5,hspace=0.5)
ax[0,0].plot(t,detuning,marker='o',markersize = ms,ls = ls,label = 'Detuning')
ax[0,0].plot(data_d[0],data_d[1],marker='o',label= 'Dynamite, detuning ',markersize=ms)
ax[0,0].plot(data_d[0],data_d[2]/2/np.pi,marker='o',label= 'Dynamite, Rabi ',markersize=ms)


ax[0,0].plot(t,rabi,marker='o',markersize = ms,ls = ls,label = 'Rabi')
ax[0,0].set_xlabel('t(us)')
ax[0,0].set_ylabel('Frequency(MHz)')
ax[0,0].legend()

ax[0,1].plot(detuning,(E[:,0])/2/np.pi,marker='o',markersize = ms,ls = ls,label= 'Ground state tenpy')
ax[0,1].set_ylabel('Energy(MHz)')
ax[0,1].set_xlabel('detuning(MHz)')
ax[0,1].legend()

ax[0,2].plot(detuning,(E[:,2]-E[:,0])/2/np.pi,marker='o',markersize = ms,ls = ls,label= 'gap')
ax[0,2].set_ylabel('Energy(MHz)')
ax[0,2].set_xlabel('detuning(MHz)')

ax[0,2].legend()

ax[1,0].plot((detuning[:-1]+detuning[1:])/2,overlap[1:,0],marker='o',markersize = ms,ls = ls)
ax[1,0].plot(data_d[1],data_d[3],marker='o',markersize=ms,ls=ls,label = 'dynamite')
ax[1,0].legend()
ax[1,0].set_xlabel('Detuning(MHz)')
ax[1,0].set_ylabel('Overlap <g^n|evolved state>')

ax[1,1].plot((detuning[:-1]+detuning[1:])/2,overlap[1:,1],marker='o',markersize = ms,ls = ls)
ax[1,1].plot(data_d[1],data_d[4],marker='o',markersize=ms,ls=ls,label = 'dynamite')
ax[1,1].legend()
ax[1,1].set_xlabel('Detuning(MHz)')
ax[1,1].set_ylabel('Overlap <AFM|evolved state>')

ax[1,2].plot(detuning,overlap[:,2]+overlap[:,3],marker='o',markersize = ms,ls = ls)
#ax[1,2].plot(detuning,overlap[:,3],marker='o',markersize = ms,ls = ls,label = 'tenpy, 1st')

ax[1,2].plot(data_d[1],data_d[5],marker='o',markersize=ms,label = 'dynamite')
ax[1,2].legend()
ax[1,2].set_xlabel('Detuning(MHz)')
ax[1,2].set_ylabel('Overlap <Ground state|evolve state>')
```
