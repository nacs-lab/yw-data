```python
import numpy as np
import os
file = '0.txt'
data = {}
data['frequency'] = []
data['power'] = []
with open(file,'r') as f:
    lines = f.readlines()[13:]
    for line in lines:
        line_1=line.split(';',1)
        data['frequency'].append(float(line_1[0]))
        data['power'].append(float(line_1[1][:-3]))
        #print(line_1[1][:-3])
```
