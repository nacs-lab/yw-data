```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from itertools import product
from scipy.optimize import curve_fit
```

```python
def generate_U(omega,detuning,dt):
  #  return np.array([[np.cos(omega*dt),-1j*np.sin(omega*dt)],[-1j*np.sin(omega*dt),np.cos(omega*dt)]])
    return np.array([[np.exp(1j * dt * detuning/2)* np.cos(dt*(detuning-omega/2)),-1j*np.exp(1j * dt * detuning/2)* np.sin(dt*(detuning-omega/2))],[-1j*np.exp(1j * dt * detuning/2)* np.sin(dt*(detuning-omega/2)),np.exp(1j * dt * detuning/2)* np.cos(dt*(detuning-omega/2))]])
```

```python
def det_func(t,det_init, det_fin, t_init,t_len):
    if t<=t_init:
        return det_init
    elif t>=t_init+t_len:
        return det_fin
    else:
        return det_init+(det_fin-det_init)*(t-t_init)/t_len
    
```

```python
det_init = 0#-4*2*np.pi
det_fin = 0#4*2*np.pi
t_init = 0
t_len = 2
T = 10
dt = 0.01
omega = 0 #1.5*2*np.pi
detuning = 0
t_list = np.arange(0,T,dt)
dim = len(t_list)
state = np.array([1.,0.])
time_trace = np.zeros_like(t_list)
det_trace = np.zeros_like(t_list)
for i in range(dim):
    det_trace[i] = det_func(t_list[i],det_init,det_fin,t_init,t_len)
    time_trace[i] = abs(state[0])**2
    state = generate_U(omega,det_trace[i],dt).dot(state)
fig, ax = plt.subplots(1, 2, figsize=(25,10))

ax[0].plot(t_list,time_trace)
ax[1].plot(t_list,det_trace)
```
