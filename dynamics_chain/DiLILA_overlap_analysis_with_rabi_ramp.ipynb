```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1_10 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/11_28/Overlap2022_11_28_12_18_04_L15_1', delimiter = ',')

data_050_10 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/11_28/Overlap2022_11_28_12_30_37_L15_1', delimiter = ',')
data0_10 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/11_28/Overlap2022_11_28_12_52_37_L15_1', delimiter = ',')

data020_10 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/11_28/Overlap2022_11_28_13_07_23_L15_1', delimiter = ',')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
#fig.suptitle('19 Atoms')
MS = 1
LS = ''
plt.subplots_adjust(wspace=0.5,hspace=0.5)

ax[0].plot(data1_10[0],data1_10[1],marker='o',label= 'Detuning = -10MHz, Rabi scan = 1us ',markersize=MS)
ax[0].plot(data0_10[0],data0_10[1],marker='o',label= 'Detuning = -10MHz, No Rabi scan ',markersize=MS)


ax[0].set_xlabel('t(us)')
ax[0].set_ylabel('Detuning(MHz)')

ax[1].plot(data1_10[0],data1_10[2]/np.pi,marker='o',markersize=MS)
ax[1].plot(data0_10[0],data0_10[2]/np.pi,marker='o',markersize=MS)

ax[1].set_ylabel('Rabi(MHz)')
ax[1].set_xlabel('t(us)')


ax[2].plot(data1_10[0],data1_10[5],marker='o',markersize=MS,ls=LS)
ax[2].plot(data0_10[0],data0_10[5],marker='o',markersize=MS)
ax[2].set_xlabel('t(us)')
ax[2].set_ylabel('Overlap with the ground state')

fig.legend()
o1 = data1_10[5][int(len(data1_10[5])/2)]
o0 = data0_10[5][int(len(data0_10[5])/2)]
print('The overlap of the evolved state with the system ground state at the critical point:')
print('With a Rabi ramp = '+str(round(o1,4)))
print('Without the Rabi ramp = '+str(round(o0,4)))
```
