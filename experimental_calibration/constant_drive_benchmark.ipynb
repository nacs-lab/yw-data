```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,14,0.0512)
```

```python
# single site
singles1 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/12_12/z_2022_12_13_01_10_34',delimiter=',')
singles30 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/12_12/z_2022_12_12_17_54_26',delimiter=',')
exp_singles1 = np.loadtxt('C:/msys64/home/wangy/experiment-control/matlab_new/analysis/scripts/12_13_22/singles1.csv',delimiter=',')
exp_singles30 = np.loadtxt('C:/msys64/home/wangy/experiment-control/matlab_new/analysis/scripts/12_13_22/singles30.csv',delimiter=',')


n1 = 1-(singles1+1)/2 
n30 = 1-(singles30+1)/2 
T = 15
end = 0.5 
decayed_n1 = (n1[:]-end)*np.exp(-t[:]/T)+end
decayed_n30 = (n30[:]-end)*np.exp(-t[:]/T)+end


fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(15,7))
#plt.plot(t,n,label= 'sim')
ax[0].plot(t,decayed_n1,label= 'sim, site 1,Rabi = 1.55MHz,T = 15us')
ax[0].errorbar(exp_singles1[0],exp_singles1[1],exp_singles1[2],fmt='o--',label = 'exp, site 1')
ax[0].legend()
ax[0].set_xlabel('Time(us)')
ax[0].set_ylabel('Survival')
ax[1].plot(t,decayed_n30,label= 'sim, site 30,Rabi = 1.50MHz,T = 15us')
ax[1].errorbar(exp_singles30[0],exp_singles30[1],exp_singles30[2],fmt='o--',label = 'exp, site 30')
ax[1].legend()
ax[1].set_xlabel('Time(us)')
ax[1].set_ylabel('Survival')
```
