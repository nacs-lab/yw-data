# 2*2 array

```python
# comparison between the numerics result and the experimental data
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from itertools import product
from scipy.optimize import curve_fit
import matplotlib.colors as colors
```

```python
data_exp = np.loadtxt('./rydberg_data/8_30_small_system/exp/2_2_-4_10.csv',delimiter=',')
data = {}
for params in [(2,2,0.0,-4,10,0.743,2.,0.05,0)]:
    data[params] = {}
    data[params]['z'] = np.loadtxt("./rydberg_data/8_30_small_system/sim/z_L%d_%d_den%.3f_delta%.1f_%.1f_omega%.3f_ramptime%.3f_dt%.3f_No%d.csv"%params,delimiter=',')
params = (2,2,0.0,-4,10,0.743,2.,0.05,0)
dt = params[7]
temp = data[params]['z']
fig,ax = plt.subplots(ncols=2,nrows=2,figsize=(7,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
x = np.arange(0,65,5)*0.0512
for i in range(4):
    #print(temp[:,i])
    ax[int(i/2),i%2].errorbar(x,data_exp[i,:],yerr=data_exp[i+4,:],ls='--',marker='.',label='experiment')
    ax[int(i/2),i%2].plot(np.arange(len(temp[:,i]))*dt,1-(temp[:,i]+1)/2,ls='-',label='numerics')
    #plt.legend(handlelength=0.5,ncol=3)
    #plt.text(20,0.75,'size=3*5',fontsize=25)
    ax[int(i/2),i%2].set_ylim([0,1])
    ax[int(i/2),i%2].set_xticks([0,1,2,3])
    ax[int(i/2),i%2].set_yticks([0,0.2,0.4,0.6,0.8,1])

    #ax[i/3,i%3].set_xlim([0,1])
    ax[int(i/2),i%2].grid()

    ax[int(i/2),i%2].set_xlabel('Time(us)')
    ax[int(i/2),i%2].set_ylabel('G survival')
ax[1,1].legend(handlelength=0.5)

plt.show()
```
