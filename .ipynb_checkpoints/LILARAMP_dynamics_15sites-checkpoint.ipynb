```python
# comparison between the numerics result and the experimental data
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from itertools import product
from scipy.optimize import curve_fit
import matplotlib.colors as colors
```

```python
# generate DILILA ramp
def DiLILA(t, t_len, dinit,gapinit, gapcrit, gapfin, dcrit, dfin, tinflect):
    numstep1 = gapinit* dcrit* t + gapcrit* dinit* (tinflect - t)
    denomstep1 = gapinit* t + gapcrit* (tinflect - t)
    numstep2 = gapfin* dcrit* (t_len-t) + gapcrit* dfin* (t-tinflect);
    denomstep2 = gapfin* (t_len-t) + gapcrit* (t-tinflect)
    if t>t_len:
        out = dfin
    elif t>tinflect:
        out = numstep2/ denomstep2
    else:
        out = numstep1/ denomstep1
    return out
```

```python
data1 = np.loadtxt('./rydberg_data/11_16/z_2022_11_17_11_55_54',delimiter=',')
data2 = np.loadtxt('./rydberg_data/11_16/z_2022_11_17_17_03_56',delimiter=',')
data3 = np.loadtxt('./rydberg_data/11_16/z_2022_11_17_16_55_43',delimiter=',')
data4 = np.loadtxt('./rydberg_data/11_16/z_2022_11_17_20_16_41',delimiter=',')
```

```python
dt = 0.0512
N = 15
fig,ax = plt.subplots(ncols=3,nrows=5,figsize=(10,10))
plt.subplots_adjust(wspace=0.5,hspace=0.1,top=0.95)
fig.suptitle('15 Atoms in a chain, standard: gapcrit,0.3,Rabi,1.2,dcrit,1.2')
for i in range(N):
    ax[int(i/3),i%3].plot(np.arange(len(data1[:,i]))*dt,1-(data1[:,i]+1)/2,ls='-',label='Standard ')
    ax[int(i/3),i%3].plot(np.arange(len(data2[:,i]))*dt,1-(data2[:,i]+1)/2,ls='-',label='gapcrit,0.35')
    ax[int(i/3),i%3].plot(np.arange(len(data3[:,i]))*dt,1-(data3[:,i]+1)/2,ls='-',label='Rabi,1.1')
    ax[int(i/3),i%3].plot(np.arange(len(data4[:,i]))*dt,1-(data4[:,i]+1)/2,ls='-',label='dcrit,1.1')

    ax[int(i/3),i%3].set_xlabel('Time(us)')
    ax[int(i/3),i%3].set_ylabel('survival of site ' + str(i+1))
ax[0,0].legend(handlelength=0.5)
```
