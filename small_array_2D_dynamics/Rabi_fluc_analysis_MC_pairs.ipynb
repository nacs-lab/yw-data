```python
import matplotlib.pyplot as plt
import glob
import numpy as np
from scipy import signal
```

```python
# #file_name = '2023_01_12_02_03_27_L2_1'
# file_name = '2023_01_12_02_38_15_L2_1'  # without interaction
# #file_name = '2023_01_12_10_23_49_L2_1'# blockaded 
# file_name = '2023_01_12_14_55_26_L2_1' # noise with omega frequency with interaction
# file_name = '2023_01_12_22_32_17_L2_1' # noise with omega frequency without interaction

#file_name = '2023_01_12_22_42_00_L2_1' # noise with sqrt(2) omega frequency without interaction
file_name = '2023_01_12_22_53_14_L2_1' # noise with sqrt(2) omega freq with interaction
data = []
z = []
data_path = glob.glob(('C:/Users/wangy/Rydberg simulation/rydberg_data/1_11_23/random_rabi_*_'+file_name))
z_path = glob.glob('C:/Users/wangy/Rydberg simulation/rydberg_data/1_11_23/z_*_'+file_name)
realizations = len(data_path)
print(realizations)
for i in range(realizations):
    data.append(np.loadtxt(data_path[i], delimiter = ','))
    z.append(np.loadtxt(z_path[i], delimiter = ','))

t = data[0][0]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].set_ylabel('Detuning(MHz)')
ax[0].set_xlabel('t(us)')
ax[1].set_xlabel('t(us)')
ax[1].set_ylabel('Rabi(MHz)')

for i in range(3):
    ax[0].plot(data[i][0][:-1], data[i][1][:-1],marker='.')

    ax[1].plot(data[i][0][:-1], data[i][2][:-1],marker='.')
    
```

```python
Nsites = np.shape(z)[-1]
fig,ax = plt.subplots(ncols= Nsites,nrows=1,figsize=(10,5))
temp = np.zeros(np.shape(z)[1])
for j in range(realizations):
    temp += 1-(z[j][:,0]+1)/2
    ax[0].plot(data[0][0],1-(z[j][:,0]+1)/2)
ax[0].set_ylabel('Ground population for each realizations')
ax[1].plot(data[0][0],temp/realizations)
ax[1].set_ylabel('Average Ground population')
ax[0].set_xlabel('t(us)')
ax[1].set_xlabel('t(us)')
# ax[0].set_ylim([0,1])
# ax[1].set_ylim([0,1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
x = temp/realizations
peaks, properties = find_peaks(x, prominence=0)
plt.scatter((data[0][0])[peaks],x[peaks])
plt.plot(data[0][0],temp/realizations,color = 'C2')
#print('Contrast for each peak = ',1-(1-x[peaks])*4)
print('Contrast for each peak = ',1-(1-x[peaks])*2)
plt.xlabel('t(us)')
# plt.plot(peaks, x[peaks], "o")
# plt.title("scipy.signal.find_peaks_cwt")
```
