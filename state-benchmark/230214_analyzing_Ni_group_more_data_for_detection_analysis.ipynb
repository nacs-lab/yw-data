```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
```

```python
def base2(n,L):
    nums = []
    while len(nums)<L:
        n, r = divmod(n, 2)
        nums.append(r)
    return nums

def undo_base2(b):
    L = len(b)
    bb = b
    return np.sum(bb*2**np.arange(L))

def is_rydberg(ar):
    return not (sum((ar*np.roll(ar,1))[1:]))
def is_rydberg_pbc(ar):
    return not (sum((ar*np.roll(ar,1))))
```

# 24 atom PBC

```python
mat = scipy.io.loadmat('data_2_9.mat')
lengths24 = [len(mat['Logics'][0][0][0]),len(mat['Logics'][0][1][0]),len(mat['Logics'][0][2][0])]
## jj = 0,1,2 for 3 different time points
posns_list = [] #list of positions of blockade violations
for jj in [0,1,2]:
    posns = []
    for j in range(lengths24[jj]):
        ar = 1-mat['Logics'][0][jj][0][j][0]
        k = undo_base2(ar)
        if not is_rydberg_pbc(ar):
            #saving the positions of the bonds of the rydberg blockade violation
            posns.append(np.nonzero((ar*np.roll(ar,1)))[0])
    #collecting an overall list of positions
    posns_list.append(np.concatenate(posns))

#frequencies of blockade violation at every bond
nums_lst = [np.histogram(p,bins=range(25))[0] for p in posns_list]
```

```python
for jj in range(3):
    plt.errorbar(range(24),nums_lst[jj]/lengths24[jj],np.sqrt(nums_lst[jj])/lengths24[jj],fmt=".")
plt.plot(range(24),np.sum(nums_lst,axis=0)/sum(lengths24),c='k',ls='dashed')
plt.ylabel("Frequency")
plt.xlabel("Bond")
plt.xticks([0,5,10,15,20],labels=["24-1","5-6","10-11","15-16","20-21"])
plt.show()
```

# Comparing to L=8 half chains

```python
mat = scipy.io.loadmat('data_2_8.mat')

lengths8 = [len(mat['Logics'][0][0][0]),len(mat['Logics'][0][1][0]),len(mat['Logics'][0][2][0])]

posnsL_list = []
posnsR_list = []
for jj in [0,1,2]:
    ryds = 0
    posnsL = []
    posnsR = []
    for j in range(lengths8[jj]):
        ##left chain
        ar = 1-mat['Logics'][0][jj][0][j][0][:8]
        k = undo_base2(ar)
        if not is_rydberg(ar):
            posnsL.append(np.nonzero((ar*np.roll(ar,1))[1:])[0])
        
        ##right chain
        ar = 1-mat['Logics'][0][jj][0][j][0][8:]
        k = undo_base2(ar)
        if not is_rydberg(ar):
            posnsR.append(np.nonzero((ar*np.roll(ar,1))[1:])[0])
    posnsL_list.append(np.concatenate(posnsL))
    posnsR_list.append(np.concatenate(posnsR))
    #print(ryds/lengths[jj])

numsL_lst = [np.histogram(p,bins=range(8))[0] for p in posnsL_list]
numsR_lst = [np.histogram(p,bins=range(8))[0] for p in posnsR_list]
```

```python
#the L=24 data
for jj in range(3):
    plt.errorbar(range(24),nums_lst[jj]/lengths24[jj],np.sqrt(nums_lst[jj])/lengths24[jj],fmt=".")
plt.plot(range(24),np.sum(nums_lst,axis=0)/sum(lengths24),c='k',ls='dashed')
#superposing the L=8 data
plt.plot(range(7),np.sum(numsL_lst,axis=0)/sum(lengths8),c='r')
plt.plot(range(13,20),np.sum(numsR_lst,axis=0)/sum(lengths8),c='r')
plt.ylabel("Frequency")
plt.xlabel("Bond")
plt.xticks([0,5,10,15,20],labels=["24-1","5-6","10-11","15-16","20-21"])
plt.show()
```
