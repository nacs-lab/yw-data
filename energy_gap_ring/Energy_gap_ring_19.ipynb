```python
import numpy as np
import matplotlib.pyplot as plt
data_10 = np.loadtxt('C:/Users/wangy/Rydberg simulation/rydberg_data/12_16/gap2022_12_16_02_25_52_ring_L19_114', delimiter = ',')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
fig.suptitle('19 Atoms')
plt.subplots_adjust(wspace=0.5,hspace=0.5)
ax[2].plot(data_10[1],data_10[5],marker='o',label='V = 10')


ax[2].legend()


ax[2].set_xlabel('Detuning(MHz)')
ax[2].set_ylabel('Overlap <AFM|ground state>')


ax[1].plot(data_10[1], (data_10[3]-data_10[2])/2/np.pi,marker='o',label='V = 10')

ax[1].set_xlabel('Detuning(MHz)')
#ax[1].set_xlim([-1,4])
ax[1].set_ylabel('Energy gap(MHz)')
ax[1].legend()
ax[0].plot(data_10[1], data_10[4],marker='o',label='V=10')
ax[0].set_xlabel('Detuning(MHz)')
ax[0].set_ylabel('Overlap <g^n|ground state>')
ax[0].legend()
ax[0].grid()
```

```python
#    op_list1 = ['t','Detuning','E0','E1','overlap_g','overlap_afm','E2','E3']
ms = 2
ls = '-'
fig, ax = plt.subplots(2, 3, figsize=(15,10))
fig.suptitle('19 Atoms')
plt.subplots_adjust(wspace=0.5,hspace=0.5)
ax[0,0].plot(data_10[0],data_10[1],marker='o',markersize = ms,ls = ls)
ax[0,0].set_xlabel('t(us)')
ax[0,0].set_ylabel('detuning(MHz)')

ax[0,1].plot(data_10[1],data_10[2]/2/np.pi,marker='o',markersize = ms,ls = ls,label= 'Ground state')
ax[0,1].set_ylabel('Energy(MHz)')
ax[0,1].set_xlabel('detuning(MHz)')
ax[0,1].legend()

ax[0,2].plot(data_10[1],data_10[3]/2/np.pi-data_10[2]/2/np.pi,marker='o',markersize = ms,ls = ls,label= '1st state gap')
ax[0,2].plot(data_10[1],data_10[6]/2/np.pi-data_10[2]/2/np.pi,marker='o',markersize = ms,ls = ls,label= '2nd state gap')
ax[0,2].plot(data_10[1],data_10[7]/2/np.pi-data_10[2]/2/np.pi,marker='o',markersize = ms,ls = ls,label= '3nd state gap')
ax[0,2].set_ylabel('Energy(MHz)')
ax[0,2].set_xlabel('detuning(MHz)')
ax[0,2].legend()

ax[1,0].plot(data_10[1],data_10[4],marker='o',markersize = ms,ls = ls)
ax[1,0].set_xlabel('Detuning(MHz)')
ax[1,0].set_ylabel('Overlap <g^n|ground state>')

ax[1,1].plot(data_10[1],2*data_10[5],marker='o',markersize = ms,ls = ls)
ax[1,1].set_xlabel('Detuning(MHz)')
ax[1,1].set_ylabel('Overlap <AFM|ground state>')
```
