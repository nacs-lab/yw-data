```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from itertools import product
from scipy.optimize import curve_fit
import matplotlib.colors as colors
from numpy import polyfit, poly1d
```

```python
# generate the distance distribution
mu, sigma = 1, 0.1 # mean and standard deviation, in unit of initial spacing
N = 10000 # numbers of sampling
r = np.random.normal(mu, sigma, N)
V0 = 1 # in unit of normalized interaction strength
V = V0/r**6
count, bins, ignored = plt.hist(r, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
```

```python
def V_distribution(V,delta_r):
    return V**(-7/6)*np.exp(-(V**(-1/6)-1)**2/(2*delta_r**2))/(6*np.sqrt(2*np.pi*delta_r**2))
```

```python
sigma = [0.01,0.02,0.05]
for i in range(len(sigma)):
    x = np.arange(0.01,10,0.001)
    V_d = V_distribution(x,sigma[i])
    plt.plot(x,V_d,label = str(sigma[i]))
    plt.legend()
```
